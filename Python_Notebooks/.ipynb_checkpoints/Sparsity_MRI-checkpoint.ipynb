# Sparsity of MRI Images
We try to test few MRI images and find out their sparsity. Of course, the used images are low-resolution images collected from public data-bases for educational purposes. The result however give some rough idea about the order of sparsity in practical images.

```python
import numpy as np
mri_np = np.array(mri_low.getdata())
print(mri_np.shape)
```

As you can see, using this method, we import the pixels into a 1D array. To be consistent with the 2D form of the image, we can reshape the result.

```python
mri_np = np.array(mri_low.getdata()).reshape(mri_low.size[0],mri_low.size[1])
print(mri_np.shape)
```

## Finding the sparsity

We can now reshape bach this image as a vector (1D-array) and calculate the frequency level of each intensity level.

```python
x = mri_np.reshape(mri_np.size,1)

# First, we find the pixels (0:255, as we have an 8-bits image) and the frequency of each pixel 
(pix, freq) = np.unique(np.copy(x), return_counts = True)
# Now we make an empirical distribution out of it
dist_x = np.array([pix , freq / x.size])

# and we plot it
plt.bar(dist_x[0,:], dist_x[1,:])
```

As you can see, the distribution shows a spike! This means sparsity. The spike however does not happen necessarily on 0. This is however not a problem. We can always bias our signal accordingly. Let's find the place of the spike:

```python
frequent_index = int(np.where(dist_x[1,:] == max(dist_x[1,:]))[0])
frequent_entry = float(dist_x[0, frequent_index])
print(frequent_entry)
```

We can now bias it, so that the most occuring entry be zero

```python
x_sparse = x - frequent_entry
```

The sparsity is now

```python
sparsity = np.count_nonzero(x != 0)
sparsity_factor = sparsity / x_sparse.size
print(sparsity, sparsity_factor)
```

You may see that the signal is not as sparse as expected: less than 10% This is due to the fact that a lot of pixels are simply not absoloutly black, but rather gray. This issue can be further addressed bu replacing the concept of sparsity with compressibility, as we learned in the lecture.

## Compressibility
Let us now ignore the pixels which are very close to black

```python
compress_threshold = 30
x_compress = np.copy(x)
x_compress[np.where(x_compress <= compress_threshold)[0],0] = 0
```

Let's see how does the MRI look like now

```python
image_compress = x_compress.reshape(512,512)
plt.imshow(image_compress, cmap = 'gray')
plt.colorbar()
```

Well, not very differenet! Now, let's see how does the sparsity change after compression.

```python
# We find the pixels and the frequency of compressed signal
(pix, freq) = np.unique(x_compress, return_counts = True)
# Now we make an empirical distribution out of it
dist_compress = np.array([pix , freq / x_compress.size])

# and we plot it
plt.bar(dist_compress[0,:], dist_compress[1,:])
```

Well this is now very sparse

```python
comp = np.count_nonzero(x_compress != 0)
comp_factor = comp / x_compress.size
print(comp, comp_factor)
```

Now, we have 30% sparse. Just change the threshold, and you will see that the sparsity could get down to 7% in you set the threshold around 60! This is what we talked about in the lecture. In nature, we might have not very sparse signals, but they are very compressible to sparse ones!

## Transformed signal

In practice, we do not rely on sampling of the data in the normal domain. We can in fact, transform data to another domain, where it is much sparser. We can of course get back to data the after recovery by inverting the transform. To see this fact, let's convert our signal to the Fourier domain by calculating its 2D-FFT:

```python
mri_fft = np.fft.fft2(mri_np)
```

We now look at the distribution of the frequency domain signal. Since mri_fft is a complex array, we can plot the distribution of the absolute value and the phase seperately.

```python
x_fft = mri_fft.reshape(mri_fft.size,1)
x_fft_abs = np.abs(x_fft)
x_fft_angle = np.angle(x_fft)
```

From the phase, we do not get any sparsity, as these type of signals usually have pretty uniform phase. Let's plot it

```python
# Let's plot the histogram
plt.hist(x_fft_angle, bins = 1000)
plt.show()
```

But, we could get high sparsity (compressibility) from the absolute value of the signal

```python
# Let's plot the histogram
hist = plt.hist(x_fft_abs, bins = 10000,range=(0,100000))
dist = hist[0]
bins_freq = hist[1]
plt.show()
```

This histogram shows high compressibility! Let's calculate sparsity. To this end, we peform same level of compression on the signal and then calculate the sparsity.

```python
compress_threshold_freq = int((30/256)*10000)
sparsity_fft = np.sum(dist[:compress_threshold_freq])/x_fft_abs.size
print(1 - sparsity_fft)
```

This is a large drop compared to the signal in the normal domain! Even going below that will keep the signal very sparse! Now, let's see what happens, if we get back from this compressed one to the image.

We first compress the signal into a sparse signal.

```python
x_fft_compress_abs = np.copy(x_fft_abs)
x_fft_compress_abs[np.where(x_fft_abs <= bins_freq[compress_threshold_freq + 1])[0],0] = 0
```

We can simply confirm that this is less than 10% sparse.

```python
print(np.count_nonzero(x_fft_compress_abs != 0) / x_fft_compress_abs.size)
```

Now, we send back the compressed signal to the time domain.

```python
# Make the complex-valued frequency-domain signal
euler_x_fft = np.cos(x_fft_angle) + 1j * np.sin(x_fft_angle)
x_fft_compress = x_fft_compress_abs * euler_x_fft

# Reshape it and dend it back to time (normal) dimain via 2D IFFT
mri_fft_compress = x_fft_compress.reshape(512,512)
mri_compress = np.fft.ifft2(mri_fft_compress)
```

We can now plot it. Let's first confirm that the signal is real and the imaginary part contains only numerical errors in the computation:

```python
real_power = np.linalg.norm(np.real(mri_compress), ord = 'fro')
imag_power = np.linalg.norm(np.imag(mri_compress), ord = 'fro')

print(real_power / (real_power + imag_power))
```

Well! This is the case, not surprisingly! We now plot the image

```python
# We use np.real() to get rid of numerically remained imaginary parts
plt.imshow(np.real(mri_compress), cmap = 'gray')
```

This is the time domain compression:

```python
plt.imshow(image_compress, cmap = 'gray')
plt.colorbar()
```

and this is the original mri:

```python
plt.imshow(mri, cmap = 'gray')
plt.colorbar()
```

Good! Looking at the corner pictures, we see that the frequency domain compression looks much better than the 30% sparse signal compressed in the time domain, and we have now only 10% sparse signal now! Of course, we could do much better, by using better transforms and also more intelligent compression approaches, but even with this naive approach, we are doing very good!
